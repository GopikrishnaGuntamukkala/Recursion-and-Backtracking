Given a dictionary of distinct words and an M x N board where every cell has one character. Find all possible words from the dictionary that can be formed by a sequence of adjacent characters on the board. We can move to any of 8 adjacent characters

Note: While forming a word we can move to any of the 8 adjacent cells. A cell can be used only once in one word.


Example 1:

Input: 

N = 1
 
dictionary = {"CAT"}

R = 3, C = 3

board = {{C,A,P},{A,N,D},{T,I,E}}

Output:

    CAT

Explanation: 

    C A P

    A N D

    T I E

Words we got is denoted using same color.

Example 2:

Input:
N = 4

dictionary = {"GEEKS","FOR","QUIZ","GO"}

R = 3, C = 3 

board = {{G,I,Z},{U,E,K},{Q,S,E}}

Output:
GEEKS QUIZ

Explanation: 

    G I Z

    U E K

    Q S E 

    Words we got is denoted using same color.



```python
class Solution:
    def wordBoggle(self,board,dictionary):
        """
        dic = [ g i z
                u e k
                q s e]
        word = Greeks quiz
        """
        
        res = []
        for i in range(len(dictionary)):
            word = dictionary[i]
            if self.find(board,word):
                res.append(word)
        return res
        
    def find(self,board,word):
        
        visited = [[False for i in range(len(board[0]))] for j in range(len(board))]
        def dfs(r,c,pos):
            if pos == len(word):
                return True
            if r >= 0 and c >= 0 and r < len(board) and c < len(board[0]) and visited[r][c] == False and board[r][c] == word[pos]:
                visited[r][c] = True
                if(dfs(r+1,c,pos+1) or dfs(r,c+1,pos+1) or dfs(r-1,c,pos+1) or dfs(r,c-1,pos+1) or dfs(r+1,c+1,pos+1) or dfs(r+1,c-1,pos+1) or dfs(r-1,c+1,pos+1) or dfs(r-1,c-1,pos+1)):
                    return True
                # to check that the character can be used in another word
                visited[r][c] = False
            
            return False
            
        indx = 0
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] == word[indx]:
                    if dfs(i,j,indx):
                        return True
        return False
    
N = 1
dictionary = ["CAT"]
r = 3
c = 3
board = [['C','A','P'],['A','N','D'],['T','I','E']]

obj = Solution()
obj.wordBoggle(board,dictionary)
```
